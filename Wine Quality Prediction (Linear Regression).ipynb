```python
X_data=np.array(X['tag'])
y_data=np.array(y)
```

```python
def mean(array):
    sum=0
    for i in array:
        sum+=i
    return (sum/len(array))
```

```python
def best_fit_slope(xs,ys):
    m=(((mean(xs)*mean(ys)) - mean(xs*ys)) / ((mean(xs)**2) - mean(xs**2)))
    b=mean(ys)-m*mean(xs**2)
    return m,b
```

```python
def train_test_split(xs,ys):
    test_size=0.2
    X_train=xs[:-int(test_size*len(xs))]
    X_test=xs[-int(test_size*len(xs)):]
    y_train=ys[:-int(test_size*len(ys))]
    y_test=ys[-int(test_size*len(ys)):]
    return X_train, X_test, y_train, y_test
```
